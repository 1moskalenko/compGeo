```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
pi = np.pi

# Import functions Pole and Bingham
import sys, os
sys.path.append(os.path.abspath('../functions'))
from Pole import Pole as Pole
from Bingham import Bingham as Bingham

# Read the bedding data from the Big Elk anticline
beasd = np.loadtxt(os.path.abspath('../data/ch6-1/beasd.txt'))

# Initialize poles
T = np.zeros(len(beasd))
P = np.zeros(len(beasd))

# Compute poles to bedding
for i in range(len(T)):
     T[i],P[i] = Pole(beasd[i,0]*pi/180,beasd[i,1]*pi/180,1)

# Compute cylindrical best fit
eigVec,confCone,bestFit = Bingham(T,P)

# Print best-fit fold axis
print('Best-fit fold axis: trend = {:.1f}, plunge {:.1f}'.format(eigVec[2,1]*180/pi,eigVec[2,2]*180/pi))
# Print best-fit plane 
print('Best-fit plane: strike = {:.1f}, dip {:.1f}\n'.format(bestFit[0]*180/pi,bestFit[1]*180/pi))
# Print confidence cone
print('95% elliptical confidence cones:')
print('Axis 1 (largest eigenvalue): Max = {:.1f}, Min {:.1f}'.format(confCone[0,0]*180/pi,confCone[0,1]*180/pi))
print('Axis 3 (lowest eigenvalue): Max = {:.1f}, Min {:.1f}'.format(confCone[1,0]*180/pi,confCone[1,1]*180/pi))
```
