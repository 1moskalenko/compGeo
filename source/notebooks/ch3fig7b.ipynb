```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Define Dip
dip = np.arange(0.1,90,0.1)*np.pi/180
# Define rake
r = np.arange(10,90,10)*np.pi/180
# Define maximum operator error
eo = 3*np.pi/180
for i in range(r.size):
    k = np.tan(r[i]+eo)-np.tan(r[i]);
    q = np.tan(r[i]+eo)*np.tan(r[i]);
    et = (k*np.cos(dip)/(1+q*(np.cos(dip))**2))
    # Plot et versus dip
    plt.plot(dip*180/np.pi,et*180/np.pi,'k-')
# Axes limits
plt.axis ([0, 90, 0, 11]);
# Label axes
plt.xlabel('dip δ(°)');
plt.ylabel('Maximum trend error $ε_t$(°)')
# Add grid
plt.grid()
```
